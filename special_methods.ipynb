1. **_init() – Constructor
Explanation:_**
The init() method is called automatically
whenever an object is created. It initializes
(assigns) the object's attributes.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
s1 = Student("Matthew", 30)
print(s1.name)
print(s1.age)
```

_**2. str() – String Representation**_
Explanation:
The str() method returns a user-friendly string
representation of an object. It is automatically
called when print() is used.


```python
class Student:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f"Student: {self.name}"
s = Student("Matthew")
print(s)
```

**_3. repr() – Official Representation**_
Explanation:
The repr() method returns the official or
developer-friendly representation of an object. It
is mainly used for debugging.

```python
class Student:
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return f"Student('{self.name}')"
s = Student("Daniella")
print(repr(s))
```

_**4. len() – Length_**
Explanation:
The len() method tells Python what the length of
an object is. It is called whenever len() is used.

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs
    def __len__(self):
        return len(self.songs)
p = Playlist(["Song1", "Song2", "Song3"])
print(len(p))
```

**_5. eq() – Equality Comparison_**
Explanation:
The eq() method defines how two objects are
compared using the == operator.


```python
class Student:
    def __init__(self, student_id):
        self.student_id = student_id
    def __eq__(self, other):
        return self.student_id == other.student_id
s1 = Student(101)
s2 = Student(101)
print(s1 == s2)
```

**_6. add() – Addition Operator**_
Explanation:
The add() method defines what happens when
the + operator is used on two objects.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"({self.x}, {self.y})"
v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2)
```

**_7. sub() – Subtraction Operator_**
Explanation:
The sub() method defines how objects behave
when the - operator is used.

```python
class Vector:
    def __init__(self, x):
        self.x = x
    def __sub__(self, other):
        return Vector(self.x - other.x)
    def __str__(self):
        return str(self.x)
v1 = Vector(10)
v2 = Vector(3)
print(v1 - v2)
```

**_8. mul() – Multiplication Operator**_
Explanation:
The mul() method defines how objects behave
when multiplied using the * operator.

```python
class Number:
    def __init__(self, value):
        self.value = value
    def __mul__(self, other):
        return Number(self.value * other.value)
    def __str__(self):
        return str(self.value)
n1 = Number(5)
n2 = Number(4)
print(n1 * n2)
```

**9. getitem() – Item Access**
Explanation:
The getitem() method allows an object to use
indexing ([]) like a list or tuple.

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]
    def __getitem__(self, index):
        return self.data[index]
g = Grades()
print(g[0])
print(g[1])
```

**_10. setitem() – Item Assignment_**
Explanation:
The setitem() method allows values to be
assigned using indexing ([]).

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]
    def __setitem__(self, index, value):
        self.data[index] = value
g = Grades()
g[1] = 100
print(g.data)
```

**_11. contains() – Membership Test**_
Explanation:
The contains() method defines how the in
operator works for an object.


```python
class Course:
    def __init__(self):
        self.students = ["Ella", "Sandra", "Joe"]
    def __contains__(self, student):
        return student in self.students
c = Course()
print("Joe" in c)
print("Gifty" in c)
```

**_12. call() – Callable Object_**
Explanation:
The call() method allows an object to behave
like a function.

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    def __call__(self, number):
        return number * self.factor
double = Multiplier(2)
print(double(10))
```

**_13. iter() and next() – Iterator_**
Explanation:
These methods allow an object to be used in
loops. iter() returns the iterator, while next()
returns the next value until no values remain.

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
for num in Counter(5):
    print(num)
```

**_14. bool() – Boolean Evaluation**_
Explanation:
The bool() method defines whether an object
should be considered True or False.

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    def __bool__(self):
        return self.balance > 0
account = BankAccount(500)
if account:
    print("Account is active")
```

**_15. del() – Destructor_**
Explanation:
The del() method is called when an object is
about to be destroyed. It is mainly used for
cleanup operations.

```python
class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        print("File opened")
    def __del__(self):
        print("File closed")
f = FileHandler("data.txt")
```

16. ShoppingCart Lab Exercise
Explanation:
This example combines several special methods
(init, str, len, getitem, setitem, eq, and add) to
make a shopping cart behave like a built-in
Python object.

```python
class ShoppingCart:
    def __init__(self, items):
        self.items = items
    def __str__(self):
        return f"Shopping Cart: {self.items}"
    def __len__(self):
        return len(self.items)
    def __getitem__(self, index):
        return self.items[index]
    def __setitem__(self, index, value):
        self.items[index] = value
    def __eq__(self, other):
        return self.items == other.items
    def __add__(self, other):
        return ShoppingCart(self.items + other.items)
cart1 = ShoppingCart(["Laptop", "Mouse",
"Keyboard"])
cart2 = ShoppingCart(["Laptop", "Mouse",
"Keyboard"])
print(cart1)
print(len(cart1))
print(cart1[0])
cart1[1] = "Headset"
print(cart1)
print(cart1 == cart2)
cart3 = cart1 + cart2
print(cart3)
```
